Testing on whole video

```python
video_path = '/content/Golf Putt.mp4'
video = cv2.VideoCapture(video_path)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = video.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter("ShotSight_Output.mp4", fourcc, fps, (640, 640))
```

```python
model = YOLO("/content/drive/MyDrive/ShotSight.pt")
```

```python
def is_overlap(boxA, boxB):
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interArea = max(0, xB - xA) * max(0, yB - yA)
    return interArea > 20
```

```python
while True:
    ret, frame = video.read()
    if not ret:
        break
    frame = cv2.resize(frame, (640, 640))
    results = model.predict(source=frame, conf=0.2, verbose=False)
    detections = results[0].boxes.data.cpu().numpy()
    class_names = results[0].names
    ball_boxes = []
    hole_boxes = []
    for det in detections:
        x1, y1, x2, y2, conf, cls_id = det
        label = class_names[int(cls_id)]

        if label == "golf_ball":
            ball_boxes.append([x1, y1, x2, y2])
        elif label == "hole":
            hole_boxes.append([x1, y1, x2, y2])
    ballputt = any(is_overlap(ball, hole) for ball in ball_boxes for hole in hole_boxes)
    if ballputt:
        cv2.putText(frame, "The putt is achieved!", (50, 60), cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255, 255, 255), 2)
    frame = results[0].plot()
    out.write(frame)

video.release()
out.release()
```
